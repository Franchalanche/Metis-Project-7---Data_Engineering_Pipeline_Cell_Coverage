```python
import pandas as pd
```

```python
df = pd.read_csv("df_no_nan_small.csv",parse_dates=["date"])
```

```python
df["date"].min()
```

```python
df["date"].unique()
```

```python
beginning = '2017-06-25T00:00:00.000000000'
```

```python
end = '2017-07-24T00:00:00.000000000'
```

```python
mask = (df['date'] >= beginning) & (df['date'] <= end)
```

```python
new_df = df.loc[mask]
```

```python
new_df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
fig, ax = plt.subplots()
ax.plot(np.unique(new_df["hour"]), new_df["hour"].value_counts().sort_index(), linewidth=2.0)
fig.set_size_inches(30, 10.5, forward=True)
```
